## Produto Cartesiano
Produto cartesiano é a relação de pares entre conjuntos, criando matrizes normalmente empregadas em Gráficos.

No pandas, esse tipo de gráfico é chamado de line plot.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Ao usar inline, o gráfico é exibido diretamente abaixo da célula, garantindo que uma nova janela não seja aberta.
%matplotlib inline
```

```python
# Dados
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]
```

```python
# Line Plot
plt.plot(x, y)

plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Produto Cartesiano entre dois Conjuntos")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
```

```python
plt.plot(x, y)
plt.plot(y, x)

plt.ylabel("Eixo X")
plt.ylabel("Eixo Y")

plt.title("Exibição Simultânea")
plt.show()
```

## Line Chart
Esse é um exemplo de gráfico em sinal, usamos seno para manipular cada elemento dentro do conjunto de x.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y)

plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.title("Line Chart")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.plot(x, y3, label="sin(x) * cos(x)")

plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.title("Funções Multiplas")

plt.grid(True, alpha=.3)
plt.show()
```
